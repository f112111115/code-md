```python
# 1. 設定圖片路徑，並加載圖片
image_path = '/content/349289_0.jpg'  # 在這裡指定你的圖片路徑
image = cv2.imread(image_path)

if image is None:
    raise FileNotFoundError(f"找不到圖片: {image_path}")

# 轉灰階處理（可選）
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2. 初始化 OCR 模型
ocr = PaddleOCR(use_angle_cls=True, lang='ch')

# 3. 使用 PaddleOCR 進行文字識別
result = ocr.ocr(gray, cls=True)

# 確保返回結果非空
if result is None or not isinstance(result, list):
    raise ValueError("OCR 沒有返回有效結果，請檢查圖片或模型設置")

# 4. 定義一個函數來判斷兩個文字是否在同一行
def is_same_row(box1, box2, threshold=10):
    y1_avg = (box1[0][1] + box1[1][1]) / 2  # box1 的上邊界平均 y 座標
    y2_avg = (box2[0][1] + box2[1][1]) / 2  # box2 的上邊界平均 y 座標
    return abs(y1_avg - y2_avg) < threshold  # 如果 y 座標差異在閾值範圍內，則視為同一行

# 5. 找到「儀器編號」的文字及其位置
instrument_text = None
instrument_box = None

for line in result:
    if line and isinstance(line, list) and len(line) > 1 and line[1]:
        text = line[1][0]  # 取得識別到的文字
        if "儀器編號" in text:  # 假設我們要找的關鍵字是 "儀器編號"
            instrument_text = text
            instrument_box = line[0]  # 儀器編號的座標框
            break

# 6. 找到同一行的其他文字
if instrument_box:
    same_row_texts = []
    for line in result:
        if line and isinstance(line, list) and len(line) > 1 and line[1]:
            box = line[0]  # 每個文字的邊界框
            text = line[1][0]  # 每個文字的內容

            if is_same_row(instrument_box, box):  # 如果這個文字和儀器編號在同一行
                same_row_texts.append(text)

    print("儀器編號所在行的文字：", same_row_texts)
else:
    print("未找到 '儀器編號'")

# 7. 可視化結果：框出「儀器編號」所在行的文字
if instrument_box:
    boxes = [line[0] for line in result if line and isinstance(line, list) and len(line) > 1 and line[1] and is_same_row(instrument_box, line[0])]  # 同一行的文字框
    txts = [line[1][0] for line in result if line and isinstance(line, list) and len(line) > 1 and line[1] and is_same_row(instrument_box, line[0])]  # 同一行的文字

    # 繪製邊界框和識別文字
    im_show = draw_ocr(image, boxes, txts, scores=None, font_path='path_to_font.ttf')
    im_show = cv2.cvtColor(im_show, cv2.COLOR_BGR2RGB)

    # 顯示結果
    plt.figure(figsize=(10,10))
    plt.imshow(im_show)
    plt.show()
else:
    print("未找到 '儀器編號'，無法顯示結果")
```

```python
from paddleocr import PaddleOCR, draw_ocr
import cv2
from PIL import Image
import numpy as np

# 定义 is_same_row 函数
def is_same_row(box1, box2, tolerance=10):
    """
    判断两个文本框是否在同一行
    参数:
    - box1: 第一个文本框的坐标 [左上角(x1, y1), 右上角(x2, y2), ...]
    - box2: 第二个文本框的坐标
    - tolerance: y 轴方向的容差范围，默认是 10 像素

    返回:
    - 布尔值，表示两个文本框是否在同一行
    """
    # 获取两个文本框的 y 轴中心点
    y1_center = (box1[0][1] + box1[2][1]) / 2
    y2_center = (box2[0][1] + box2[2][1]) / 2
    # 比较 y 轴中心点，判断是否在同一行
    return abs(y1_center - y2_center) < tolerance

# 初始化 PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='ch')

# 加载图片
img_path = '/content/349291_0.jpg'
image = cv2.imread(img_path)

# OCR 识别
result = ocr.ocr(img_path, cls=True)

# 将结果按行保存
boxes = []
texts = []
scores = []

# 找到“儀器编號”及对应的号碼
for line in result:
    for word_info in line:
        text = word_info[1][0]  # 获取识别的文本
        if "儀器编號" in text:
            instr_box = word_info[0]  # 获取 "儀器编號" 的位置
            boxes.append(instr_box)
            texts.append(text)
            scores.append(word_info[1][1])  # 置信度

            # 寻找同一行的其他文字
            for other_word_info in line:
                other_text = other_word_info[1][0]
                other_box = other_word_info[0]

                # 判断是否是同一行且不是 "儀器编號" 本身
                if other_word_info != word_info and is_same_row(instr_box, other_box):
                    boxes.append(other_box)
                    texts.append(other_text)
                    scores.append(other_word_info[1][1])

# 使用 draw_ocr 将识别结果画出来，指定字体路径
font_path = './NotoSansCJKsc-Regular.otf'  # 下载字体后的路径
image = Image.open(img_path)
image_with_boxes = draw_ocr(image, boxes, texts, scores, font_path=font_path)

# 将 numpy.ndarray 转换为 PIL.Image
image_with_boxes = Image.fromarray(np.uint8(image_with_boxes))

# 显示图像
image_with_boxes.show()

# 或者保存图像
image_with_boxes.save('result_with_boxes.jpg')
```
