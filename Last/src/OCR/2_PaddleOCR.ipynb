```python
import os
from paddleocr import PaddleOCR, draw_ocr
from PIL import Image
import cv2

# 初始化PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='ch')

# 设置路径
img_folder = '/content/drive/MyDrive/高科手機AI辨識/New_material/cut'  # 原始图片文件夹
output_folder = '/content/drive/MyDrive/高科手機AI辨識/New_material/New_Output'  # 处理后图片的输出文件夹

# 检查并创建输出文件夹
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 支持的文件扩展名
valid_extensions = ('.jpg', '.jpeg', '.png')

# 遍历img_folder中的所有图片
for img_name in os.listdir(img_folder):
    if img_name.lower().endswith(valid_extensions):  # 检查文件扩展名
        img_path = os.path.join(img_folder, img_name)

        # 进行OCR识别
        result = ocr.ocr(img_path, cls=True)

        # 检查OCR结果是否为None或结果为空
        if result is None or not result or not result[0]:
            print(f"No valid OCR result for {img_name}")
            continue  # 跳过没有OCR结果的图片

        res = result[0]

        # 打开图片
        image = Image.open(img_path).convert('RGB')

        # 获取识别的文字、框以及置信度
        boxes = [line[0] for line in res if line]
        txts = [line[1][0] for line in res if line]
        scores = [line[1][1] for line in res if line]

        # 绘制标记在图片上
        im_show = draw_ocr(image, boxes, txts, scores, font_path='./TaipeiSansTCBeta-Regular.ttf')
        im_show = Image.fromarray(im_show)

        # 保存到新的文件夹
        output_image_path = os.path.join(output_folder, img_name)
        im_show.save(output_image_path)

        print(f"Processed and saved: {output_image_path}")
```

```python
import cv2
import pytesseract
import re
import os

# 图片存放路径
image_folder = "/content/drive/MyDrive/高科手機AI辨識/New_material/New_Output"
output_folder = "/content/drive/MyDrive/高科手機AI辨識/New_material/New_Output_txt"

# 创建输出文件夹
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 定义正则表达式
pattern = r'[A-Za-z]\d{1,2}-\d{3,}'

# 获取所有图片文件
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

# 遍历所有图片文件
for image_file in image_files:
    # 读取图片
    image_path = os.path.join(image_folder, image_file)
    image = cv2.imread(image_path)

    # 使用pytesseract提取图片中的文本
    text = pytesseract.image_to_string(image)

    # 匹配正则表达式
    matches = re.findall(pattern, text)

    # 将结果写入txt文件
    if matches:
        # 将图片文件名称转换为txt文件名称
        output_file_name = os.path.splitext(image_file)[0] + '.txt'
        output_file = os.path.join(output_folder, output_file_name)

        # 检查输出文件的目录是否存在
        output_dir = os.path.dirname(output_file)
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        with open(output_file, 'w', encoding='utf-8') as f:  # 加上 encoding='utf-8' 支援中文
            f.write(f'Image: {image_file}\n')
            f.write('Detected Text:\n')
            for match in matches:
                f.write(f'{match}\n')
            f.write('-' * 50 + '\n')

print("处理完成！所有符合条件的文本已保存为 .txt 文件。")
```
